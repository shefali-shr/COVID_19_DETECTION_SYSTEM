```python
import pandas as pd
import os
import shutil
```

```python
File_path="covid-chestxray-dataset/metadata.csv"
Image_path="covid-chestxray-dataset/images"
```

```python
df=pd.read_csv(File_path)
```

```python
df.head()
```

```python
df.shape
```

```python
target_directory="dataset/covid"

if not os.path.exists(target_directory):
    os.mkdir(target_directory)
```

```python
df["finding"].value_counts()
```

```python
df['finding'].replace({'Pneumonia/Viral/COVID-19': 'COVID-19'},inplace=True)
```

```python
df["finding"].value_counts()
```

```python
count=0
for(i,row) in df.iterrows():
    if row['finding']=="COVID-19":
        count+=1
        
print(count)
```

```python
df.columns
```

```python
df['view']
```

```python
count=0
for(i,row) in df.iterrows():
    if row['finding']=="COVID-19" and row['view']=="PA":
        count+=1
        
print(count)
```

```python
count=0
for(i,row) in df.iterrows():
    if row['finding']=="COVID-19" and row['view']=="PA":
        filename=row["filename"]
        
        
        image_path=os.path.join(Image_path,filename)
        image_copy_path=os.path.join(target_directory,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving images")
        count+=1
        
print(count)
```

```python
#sampling for kaggle images

import random
kaggle_file_path="chest_xray/train/NORMAL"
target_normal_dir="dataset/normal_images"
```

```python
image_names=os.listdir(kaggle_file_path)
```

```python
random.shuffle(image_names)
```

```python
for i in range(196):
    img_name=image_names[i]
    image_path=os.path.join(kaggle_file_path,img_name)
    target_path=os.path.join(target_normal_dir,img_name)
    shutil.copy2(image_path,target_path)
    print("copying image",i)
```
